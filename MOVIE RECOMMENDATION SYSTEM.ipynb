```python
def top_3_actor(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter !=3:
            L.append(i['name'])
            counter+=1
        else:
                break
    return L   
```

```python
#>> From the genres, keywords,cast,crew column we need to remove the space in between the words in order avoid the confusion to
#> the recommender system .As we know that we are combining the group of list together to form the tags .If we didnt remove 
#>the space then their is a chance of misinterpreting the need of the user.
#>EX:- If i want to search sam worthington movie if i didnt remove space then other person name which starts from sam will be shown.
```

```python
#>> After fetching the common 5000 words we have 1 problem that is many words have same root word.
#>In order to fetch only the root word from the multiple same word we will do stemming .
#>[loved,loving,lovers,lover]
#>stem=[love]
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
#> user defined function to fetch the root word from the tags
```

```python
def stem(text):
    L = []
    
    for i in text.split():
        L.append(ps.stem(i))
    return " ".join(L)
```

```python
print(stem("loving"))
print(stem("loved"))
print(stem("lover"))
```

```python
#>> user defined function for fetching the top 5 related movies to the input movie
```

```python
def recommend(movie):
    movie_index=new_df[new_df['title']==movie].index[0]
    distance=similarity[movie_index]
    movies_list=sorted(list(enumerate(distance)),reverse=True,key=lambda x:x[1]) [1:6] 
    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend("The Dark Knight Rises")
```
